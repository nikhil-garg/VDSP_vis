```python
data = pd.read_csv('learning_rate_50.csv')
lr = np.array(data['lr'])
iterations = np.array(data['iterations'])
accuracy = np.array(data['accuracy'])
seed = np.array(data['seed'])

index_1 = np.where(iterations==1)
index_2 = np.where(iterations==2)
index_3 = np.where(iterations==3)
index_4 = np.where(iterations==4)
index_5 = np.where(iterations==5)

lr_1 = lr[index_1]
lr_2 = lr[index_2]
lr_3 = lr[index_3]
lr_4 = lr[index_4]
lr_5 = lr[index_5]

accuracy_1 = accuracy[index_1]
accuracy_2 = accuracy[index_2]
accuracy_3 = accuracy[index_3]
accuracy_4 = accuracy[index_4]
accuracy_5 = accuracy[index_5]

seed_1 = seed[index_1]
seed_2 = seed[index_2]
seed_3 = seed[index_3]
seed_4 = seed[index_4]
seed_5 = seed[index_5]

lr_unique_1 = np.unique(lr_1)
accuracy_std_temp = np.zeros((len(lr_unique_1),5))
accuracy_mean_1 = np.zeros(len(lr_unique_1))
accuracy_std_1 = np.zeros(len(lr_unique_1))
temp = 0
for i in range(len(lr_unique_1)):
    for j in range(len(lr_1)):
         if(lr_unique_1[i] == lr_1[j]):
                accuracy_mean_1[i]+=accuracy_1[j]
                accuracy_std_temp[i][temp]=accuracy_1[j]
                temp+=1
    temp=0
                
accuracy_mean_1 /=5
accuracy_std_1 = np.std(accuracy_std_temp,1)

lr_unique_2 = np.unique(lr_2)
accuracy_std_temp = np.zeros((len(lr_unique_2),5))
accuracy_mean_2 = np.zeros(len(lr_unique_2))
for i in range(len(lr_unique_2)):
    for j in range(len(lr_2)):
         if(lr_unique_2[i] == lr_2[j]):
                accuracy_mean_2[i]+=accuracy_2[j]
                accuracy_std_temp[i][temp]=accuracy_2[j]
                temp+=1
    temp=0
accuracy_mean_2 /=5
accuracy_std_2 = np.std(accuracy_std_temp,1)

lr_unique_3 = np.unique(lr_3)
accuracy_std_temp = np.zeros((len(lr_unique_3),5))
accuracy_mean_3 = np.zeros(len(lr_unique_3))
temp = 0
for i in range(len(lr_unique_3)):
    for j in range(len(lr_3)):
         if(lr_unique_3[i] == lr_3[j]):
                accuracy_mean_3[i]+=accuracy_3[j]
                accuracy_std_temp[i][temp]=accuracy_3[j]
                temp+=1
    temp = 0
accuracy_mean_3 /=5
accuracy_std_3 = np.std(accuracy_std_temp,1)

lr_unique_4 = np.unique(lr_4)
accuracy_std_temp = np.zeros((len(lr_unique_4),5))
accuracy_mean_4 = np.zeros(len(lr_unique_4))
temp = 0
for i in range(len(lr_unique_4)):
    for j in range(len(lr_4)):
         if(lr_unique_4[i] == lr_4[j]):
                accuracy_mean_4[i]+=accuracy_4[j]
                accuracy_std_temp[i][temp]=accuracy_4[j]
                temp+=1
    temp = 0
accuracy_mean_4 /=5
accuracy_std_4 = np.std(accuracy_std_temp,1)

lr_unique_5 = np.unique(lr_5)
accuracy_std_temp = np.zeros((len(lr_unique_5),5))
accuracy_mean_5 = np.zeros(len(lr_unique_5))
temp = 0
for i in range(len(lr_unique_5)):
    for j in range(len(lr_5)):
         if(lr_unique_5[i] == lr_5[j]):
                accuracy_mean_5[i]+=accuracy_5[j]
                accuracy_std_temp[i][temp]=accuracy_5[j]
                temp+=1
    temp = 0
accuracy_mean_5 /=5
accuracy_std_5 = np.std(accuracy_std_temp,1)
```

```python
f, ax = plt.subplots(figsize=(7, 4))
import matplotlib as mpl
mpl.rcParams.update({'font.size': 14})
ax.set_ylim(15,85)
ax.grid(which='major', linewidth=0.8)
ax.grid(which='minor', linestyle=':', linewidth=0.5)
ax.minorticks_on()
plt.xscale("log")  
plt.xlabel("Learning rate")
plt.ylabel("Accuracy (%)")

plt.errorbar(lr_unique_1, accuracy_mean_1,yerr=accuracy_std_1,fmt='-*',linewidth = 1.5, alpha = 0.66, capsize = 8,label='Epochs = 1' )
plt.errorbar(lr_unique_2, accuracy_mean_2,yerr=accuracy_std_2,fmt='-x',linewidth = 1.5, alpha = 0.66, capsize = 8,label='Epochs = 2' )
plt.errorbar(lr_unique_3, accuracy_mean_3,yerr=accuracy_std_3,fmt='-+',linewidth = 1.5, alpha = 0.66, capsize = 8,label='Epochs = 3' )
plt.errorbar(lr_unique_4, accuracy_mean_4,yerr=accuracy_std_4,fmt='-s',linewidth = 1.5, alpha = 0.66, capsize = 8,label='Epochs = 4' )
plt.errorbar(lr_unique_5, accuracy_mean_5,yerr=accuracy_std_5,fmt='-1',linewidth = 1.5, alpha = 0.66, capsize = 8,label='Epochs = 5' )

# plt.gca().set_title('Iterations')
plt.legend()
plt.tight_layout()

# plt.show()

plt.savefig('lr_combined_error.pdf',dpi=500)
plt.savefig('lr_combined_error.svg',dpi=500)
plt.savefig('lr_combined_error.jpeg',dpi=500)
```
